```python
groceries_df = pd.DataFrame(pd.Series(groceries_list))
groceries_df.head(10)
```

```python
groceries_df.columns = ["transactions"]
```

```python
X = groceries_df["transactions"].str.join(sep='*').str.get_dummies(sep="*")
X
```

```python
#Building rules using apriori algorithm
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
groceries_rules = apriori(X, min_support=0.01, max_len=3,use_colnames=True)
groceries_rules
```

```python
groceries_rules.sort_values('support',ascending=False,inplace=True)
groceries_rules
```

```python
#Let's try and plot the top rules
```

```python
groceries_rules['itemsets_plot'] = groceries_rules['itemsets'].apply(lambda x:", ".join(list(x))).astype("unicode")
groceries_rules
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,5))
plt.barh(y=range(0,11),width=groceries_rules.support[0:11],color="rgby");plt.xlabel('support');plt.ylabel('items/rules')
#annotating the bar plot
x = np.array(groceries_rules.support[0:11])
y = range(0,11)
i=0
for x,y in zip(x,y):
    while(i<len(groceries_rules.index)):
        plt.annotate(groceries_rules.itemsets_plot[groceries_rules.index[i]],(x,y),textcoords="offset points",xytext=(0,10))
        i+=1
        break
plt.show()  
```

```python
#Now let's build all rules using association
```

```python
groceries_rules_all = association_rules(groceries_rules.iloc[:,:2],metric='lift',min_threshold=1)
```

```python
groceries_rules_all.sort_values('lift',ascending=False,inplace=True)
groceries_rules_all
```

```python
#Let's plot the top rules
```

```python
groceries_rules_all['antecedents'] = groceries_rules_all['antecedents'].apply(lambda x:", ".join(list(x))).astype("unicode")
groceries_rules_all['consequents'] = groceries_rules_all['consequents'].apply(lambda x:", ".join(list(x))).astype("unicode")
groceries_rules_all
```

```python
#Putting together antecendent & consequent for annotations
ant_cons = []
for i in groceries_rules_all.index:
    while(i<len(groceries_rules_all.antecedents) and i<len(groceries_rules_all.consequents)):
        ant_cons.append("("+groceries_rules_all.antecedents[i] +") & ("+ groceries_rules_all.consequents[i]+")")
        break
ant_cons[:11]
```

```python
plt.figure(figsize=(15,7))
plt.barh(y=range(0,11), width =groceries_rules_all.lift[0:11], color="rgby");plt.ylabel('Association rules');plt.xlabel('lift');plt.xticks(range(0,11))
#adding annotations to the bar plot
x = np.array(groceries_rules_all.lift[0:11])
y = range(0,11)
i = 0
for x,y in zip(x,y):
    while(i<11):
        plt.annotate(ant_cons[i],(x,y),textcoords="offset points",xytext=(0,10))
        i+=1
        break;
plt.show()
```
